```python
import numpy as np
```

```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open ('losses/emloss', 'rb') as fp:
    em = pickle.load(fp)
with open ('losses/iemloss', 'rb') as fp:
    iem = pickle.load(fp)
with open ('losses/oemloss', 'rb') as fp:!
    oem = pickle.load(fp)
with open ('losses/oemvrloss', 'rb') as fp:
    oemvr = pickle.load(fp)
```

```python
epochs = len(em)
```

# With 1000 documents

```python
with open ('losses/emlossbig', 'rb') as fp:
    em = pickle.load(fp)
with open ('losses/iemlossbig', 'rb') as fp:
    iem = pickle.load(fp)
with open ('losses/oemlossbig', 'rb') as fp:
    oem = pickle.load(fp)
with open ('losses/oemvrlossbig', 'rb') as fp:
    oemvr = pickle.load(fp)
```

```python
plt.plot(np.arange(epochs), em, label='EM')
plt.plot(np.arange(epochs), iem, label='IEM')
leg = plt.legend(fontsize=20,fancybox=True, loc='right')
leg.get_frame().set_alpha(0.5)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('Objective', fontsize=15)
plt.show()
```

```python
plt.plot(np.arange(epochs), em, label='EM')
plt.plot(np.arange(epochs), iem, label='IEM')
plt.plot(np.arange(epochs), oem, label='OEM')
plt.plot(np.arange(epochs), oemvr, label='OEMVR')
leg = plt.legend(fontsize=20,fancybox=True, loc='right')
leg.get_frame().set_alpha(0.5)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('Objective', fontsize=15)
plt.show()
```
