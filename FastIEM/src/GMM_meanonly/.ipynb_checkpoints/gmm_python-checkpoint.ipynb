```python
def f(x, y, mu):
    z = np.column_stack([x.ravel(), y.ravel()])
    return (0.2*st.multivariate_normal([mu,2], 1*np.eye(2)).pdf(z) +
            0.8*st.multivariate_normal([2,-mu], 1*np.eye(2)).pdf(z))
```

```python
f(np.arange(3), np.arange(3), mu=4)
```

```python
s = 200
mu = 3
x = np.linspace(-3, 6, s)
y = np.linspace(-3, 8, s)
X, Y = np.meshgrid(x, y)
Z = np.reshape(f(X, Y, mu), (s, s))

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='jet')
plt.title('Gaussian Mxixture Model');
```
