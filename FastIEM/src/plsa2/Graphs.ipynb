```python
import numpy as np
```

```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open ('losses/emloss', 'rb') as fp:
    em = pickle.load(fp)
with open ('losses/iemloss', 'rb') as fp:
    iem = pickle.load(fp)
```

```python
epochs = len(em)
```

```python
plt.plot(np.arange(epochs), [-x for x in em], label='EM')
plt.plot(np.arange(epochs), [-x for x in iem], label='IEM')
leg = plt.legend(fontsize=20,fancybox=True, loc='right')
leg.get_frame().set_alpha(0.5)
plt.xlabel('Epoch', fontsize=15)
plt.ylabel('Objective', fontsize=15)
plt.show()
```
