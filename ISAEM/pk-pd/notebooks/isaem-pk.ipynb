```python
dim = "Cl"
```

```python
saemrun_1chain = list(onechain[(onechain.algo=='full') & (onechain.variable==dim)].value)
isaemrun25_1chain = list(onechain[(onechain.algo=='quarter') & (onechain.variable==dim)].value)
isaemrun50_1chain = list(onechain[(onechain.algo=='half') & (onechain.variable==dim)].value)
isaemrun75_1chain = list(onechain[(onechain.algo=='three quarter') & (onechain.variable==dim)].value)
fivechain = pd.read_csv('data/pk100_5chains.csv')
saemrun_5chain = list(fivechain[(fivechain.algo=='full') & (fivechain.variable==dim)].value)
isaemrun25_5chain = list(fivechain[(fivechain.algo=='quarter') & (fivechain.variable==dim)].value)
isaemrun50_5chain = list(fivechain[(fivechain.algo=='half') & (fivechain.variable==dim)].value)
isaemrun75_5chain = list(fivechain[(fivechain.algo=='three quarter') & (fivechain.variable==dim)].value)
tenchain = pd.read_csv('data/pk100_10chains.csv')
saemrun_10chain = list(tenchain[(tenchain.algo=='full') & (tenchain.variable==dim)].value)
isaemrun25_10chain = list(tenchain[(tenchain.algo=='quarter') & (tenchain.variable==dim)].value)
isaemrun50_10chain = list(tenchain[(tenchain.algo=='half') & (tenchain.variable==dim)].value)
isaemrun75_10chain = list(tenchain[(tenchain.algo=='three quarter') & (tenchain.variable==dim)].value)

twentychain = pd.read_csv('data/pk100_20chains.csv')
saemrun_20chain = list(twentychain[(twentychain.algo=='full') & (twentychain.variable==dim)].value)
isaemrun25_20chain = list(twentychain[(twentychain.algo=='quarter') & (twentychain.variable==dim)].value)
isaemrun50_20chain = list(twentychain[(twentychain.algo=='half') & (twentychain.variable==dim)].value)
isaemrun75_20chain = list(twentychain[(twentychain.algo=='three quarter') & (twentychain.variable==dim)].value)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(16, 4))
ax = plt.subplot(1, 3, 1)
#ax.set_yscale('log')
#ax.set_xscale('log')
plt.plot(xaxisrun, saemrun_1chain, label='SAEM', linestyle='--')
plt.plot(xaxisrun25, isaemrun25_1chain, label='SAEM 25%', linestyle='--')
plt.plot(xaxisrun50, isaemrun50_1chain, label='SAEM 50%', linestyle='--')
plt.plot(xaxisrun75, isaemrun75_1chain, label='SAEM 75%', linestyle='--')
leg = plt.legend(fontsize=14,fancybox=True, loc=0,ncol=2)
leg.get_frame().set_alpha(0.5)
plt.xlabel('Iterations', fontsize=20)
plt.ylabel('Cl', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(linestyle='dotted',linewidth=2)

ax = plt.subplot(1, 3, 2)
#ax.set_yscale('log')
#ax.set_xscale('log')
plt.plot(xaxisrun, saemrun_10chain, label='SAEM', linestyle='--')
plt.plot(xaxisrun25, isaemrun25_10chain, label='SAEM 25%', linestyle='--')
plt.plot(xaxisrun50, isaemrun50_10chain, label='SAEM 50%', linestyle='--')
plt.plot(xaxisrun75, isaemrun75_10chain, label='SAEM 75%', linestyle='--')
leg = plt.legend(fontsize=14,fancybox=True, loc=0,ncol=2)
leg.get_frame().set_alpha(0.5)
plt.xlabel('Iterations', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(linestyle='dotted',linewidth=2)

ax = plt.subplot(1, 3, 3)
#ax.set_yscale('log')
#ax.set_xscale('log')
plt.plot(xaxisrun, saemrun_20chain, label='SAEM', linestyle='--')
plt.plot(xaxisrun25, isaemrun25_20chain, label='SAEM 25%', linestyle='--')
plt.plot(xaxisrun50, isaemrun50_20chain, label='SAEM 50%', linestyle='--')
plt.plot(xaxisrun75, isaemrun75_20chain, label='SAEM 75%', linestyle='--')
leg = plt.legend(fontsize=14,fancybox=True, loc=0,ncol=2)
leg.get_frame().set_alpha(0.5)
plt.xlabel('Iterations', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(linestyle='dotted',linewidth=2)

fig.tight_layout()
plt.show()
```
